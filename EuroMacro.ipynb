```python
gdp_clean = gdp[['Geopolitical entity (reporting)', 'TIME_PERIOD', 'OBS_VALUE']]
gdp_clean.columns = ['Country', 'Year', 'GDP']
```

```python
inflation_clean = inflation[['Geopolitical entity (reporting)', 'TIME_PERIOD', 'OBS_VALUE']]
inflation_clean.columns = ['Country', 'Year', 'Inflation']
```

```python
unemployment_clean = unemployment[['Geopolitical entity (reporting)', 'TIME_PERIOD', 'OBS_VALUE']]
unemployment_clean.columns = ['Country', 'Year', 'Unemployment']
```

```python
inflation.columns
```

```python
inflation_clean.columns
```

```python
# Merging GDP and inflation btw
merged_data = gdp_clean.merge(inflation_clean, on=['Country', 'Year'], how='inner')

merged_data = merged_data.merge(unemployment_clean, on=['Country', 'Year'], how='inner')

merged_data.head()
```

```python
# Descriptive stats
stats = merged_data[['GDP', 'Inflation', 'Unemployment']].describe()
stats
```

```python
correlation = merged_data[['GDP', 'Inflation', 'Unemployment']].corr()
correlation
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
merged_data.to_csv('europe_macro_data.csv', index=False)
```
